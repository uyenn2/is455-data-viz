```python
options = df3['index']
@ipywidgets.interact (vol_op = options)
def make_pie_chart(vol_op):
    total_count = df3[df3['index'] != vol_op]['Type'].sum()
    selected_vol = df3[df3['index'] == vol_op]['Type'].values[0]
    proportions = [selected_vol, total_count]
    
    plt.figure()
    plt.pie(proportions, labels = [vol_op, total_count], autopct = '%1.1f%%', startangle = 90)
    plt.title(f'Percentages of Types')
    plt.show()
```

```python
options2 = df_country['index']
@ipywidgets.interact (country = options2)
def make_pie_chart(country):
    total_country2 = df_country[df_country['index'] != country]['Country'].sum()
    selected_country = df_country[df_country['index'] == country]['Country'].values[0]
    proportions2 = [selected_country, total_country2]
    
    plt.figure()
    plt.pie(proportions2, labels = [country, total_country2], autopct = '%1.1f%%', startangle = 90)
    plt.title(f'Percentages of Country')
    plt.show()
```
